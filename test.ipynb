```python

from IPython.html.widgets import interact
import IPython.html.widgets as widgets

def test_interact(z0, b, f,s,dmin,dmax,dstep):
    d = np.linspace(dmin, dmax, (dmax-dmin)/dstep)    
    z =(z0*b)/(b - ((z0/f)*(s/1000))*d)   
    zd =(b*s*z0*z0)/(1000*f*(b - ((d*s*z0)/(1000*f)))*(b - ((d*s*z0)/(1000*f))))
    figure(1)
    plot(d, z)
    figure(2)
    plot(d, zd)
interact(test_interact,
         z0=widgets.FloatSlider(min=200.0,max=1500.0, step=50.0, value=1300.0),\
         b =widgets.FloatSlider(min=5.0, max=100.0, step=1.0, value=75.0),\
         f =widgets.FloatSlider(min=1.0, max=6.0, step=0.01, value=5.9),\
         s =widgets.FloatSlider(min=1.12, max=6.0, step=0.01, value=5.2),\
         dmin=-128,dmax=60,dstep=0.25)    


```
